```python
#Get prices in the native currency
ticker = ["MSFT", "ISP.MI", "NESN.SW", "HSBA.L", "7203.T", "RY.TO", "CBA.AX", "0005.HK"]     # (7203.T is Toyota and 0005.HK is HSBC)
raw = dd.get_raw_prices(ticker, start = "2025-01-01")
raw.tail()
```

# Converting the prices in various currencies

```python
base = "EUR"
converted = dd.convert_to_base(raw, base = base, show_currency_detection = True)
print(f"\n\n>>> Converted Prices in {base} (last 5 rows):")
display(converted.tail())
```
